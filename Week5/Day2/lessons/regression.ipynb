# Linear vs. Logistic Regression


## Linear Regression

Linear regression is a fundamental supervised learning algorithm used for predicting continuous numerical values based on input features. The goal of linear regression is to find a linear relationship between the independent variables (features) and the dependent variable (target/output).

The model learns the values of and by minimizing the error between predicted and actual values.

Example use case: Predicting house prices based on square footage.

- Input: Square footage of a house

- Output: Predicted price of the house

- Features: Number of bedrooms, location, and square footage

How It Works:

The algorithm takes input data and tries to fit a straight line that best represents the relationship between the input and output.
The weights and bias are adjusted iteratively using an optimization method (e.g., Gradient Descent) to minimize the error.
Once trained, the model can make predictions on new data.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Example Data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])

# Model Training
model = LinearRegression()
model.fit(X, y)
predictions = model.predict(X)

# Plot
plt.scatter(X, y, color='blue')
plt.plot(X, predictions, color='red')
plt.xlabel("Input Feature")
plt.ylabel("Target")
plt.title("Linear Regression")
plt.show()
```

## Logistic regression


Logistic regression is used for classification problems, predicting discrete categories (e.g., spam vs. not spam).

```python
from sklearn.linear_model import LogisticRegression

# Example Data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([0, 0, 1, 1, 1])

# Model Training
log_model = LogisticRegression()
log_model.fit(X, y)
predictions = log_model.predict(X)
print("Predictions:", predictions)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Example Data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([0, 0, 1, 1, 1])

# Train logistic regression model
log_model = LogisticRegression()
log_model.fit(X, y)

# Generate a smooth range of X values for plotting
X_range = np.linspace(0, 6, 300).reshape(-1, 1)
y_proba = log_model.predict_proba(X_range)[:, 1]  # Probability of class 1

# Plot sigmoid curve
plt.plot(X_range, y_proba, label="Sigmoid Curve (P(y=1|x))", color="blue")
plt.scatter(X, y, c=y, cmap="bwr", edgecolor="k", s=80, label="Data Points")
plt.axhline(0.5, color="gray", linestyle="--", label="Decision Threshold (0.5)")
plt.xlabel("X")
plt.ylabel("Predicted Probability")
plt.title("Logistic Regression Probability Curve")
plt.legend()
plt.show()
```

## Loss Functions: MSE, MAE, and Cross-Entropy Loss

### Mean Squared Error (MSE)
MSE is used for regression tasks. It measures the average squared difference between predicted and actual values.

```python
from sklearn.metrics import mean_squared_error

# Example Predictions
y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]

print("MSE:", mean_squared_error(y_true, y_pred))
```

### Mean Absolute Error (MAE)
MAE is another regression loss function that calculates the average absolute difference between actual and predicted values. It is more robust to outliers compared to MSE.

```python
from sklearn.metrics import mean_absolute_error

print("MAE:", mean_absolute_error(y_true, y_pred))
```

### Cross-Entropy Loss
Used for classification problems, it measures how different the predicted probability distribution is from the actual labels.

```python
from sklearn.metrics import log_loss

# Example Data
y_true = [0, 1, 1, 0]
y_pred_probs = [0.1, 0.9, 0.8, 0.3]  # Probabilities

print("Cross-Entropy Loss:", log_loss(y_true, y_pred_probs))
```
