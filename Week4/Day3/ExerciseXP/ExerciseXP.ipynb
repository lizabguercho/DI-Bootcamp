### Exercise 2: Descriptive Statistics
Task: Given a sample dataset, calculate the mean, median, variance, and standard deviation using SciPy

```python
import numpy as np 
from scipy import stats

data = [12, 15, 13, 12, 18, 20, 22, 21]
```

```python
mean = stats.tmean(data)
median = np.median(data)
var = stats.tvar(data)
std = stats.tstd(data)

print("Mean:",mean)
print("Median:",median)
print("Variance:",var)
print("Standard Deviation:",std)
```

###  Exercise 3: Understanding Distributions
Task: Generate a normal distribution using SciPy with a mean of 50 and a standard deviation of 10. Plot this distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu, sigma = 50, 10
x = np.linspace(10, 90, 200)  
y = stats.norm.pdf(x, mu, sigma)

plt.plot(x, y)
plt.title("Normal Distribution (mean=50, std=10)")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

###  Exercise 4: T-Test Application
Task: Perform a T-test on two sets of randomly generated data.

```python
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)
```

```python
t_test,p_value = stats.ttest_ind(data1,data2)
print("T-statistic:", t_test)
print("p-value",p_value)
```

Conclusion: There is a significant difference between these two sets of data based on t-statistics = -6  and very small p_value.

### Exercise 5: Linear Regression Analysis
- Objective: Apply linear regression to a dataset and interpret the results.
- Task: Given a dataset of housing prices (house_prices) and their corresponding sizes (house_sizes), use linear regression to predict the price of a house given its size.


```python
house_sizes = [50, 70, 80, 100, 120] 
house_prices = [150000, 200000, 210000, 250000, 280000]
```

```python
house_sizes_np = np.array(house_sizes)
house_prices_np = np.array(house_prices)
print(house_sizes_np)
print(house_prices_np)
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes_np, house_prices_np)
# The slope and intercept of the regression line
print("Slope:", round(slope,2))
print("Intercept:", round(intercept,2))
```

```python
# predict price for 90m2 apartment
pred = intercept + slope *90
print(round(pred,2))
```

Slope: Each extra 1 m adds about 1,828.77 currency units to price.

### Exercise 6: Understanding ANOVA
- Objective: Test understanding of ANOVA and its application.

- Task: Three different fertilizers are applied to three separate groups of plants to test their effectiveness. The growth in centimeters is recorded.

```python
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]
```

```python
# Performing one-way ANOVA
f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print("F-value:", f_value)
print("P-value:", p_value)
```

Conclusions:
- F value is a ratio of variance between groups and variance within groups
- p-value is how likely my results are if null hypothesis is true
- Based on p-value = 2.754267209804844e-05 (very small) fertilizers have significantly different effects on plant growth.
- If p-value would be > 0.05 we will not see any significantly different effects on plant growth among all groups of fertilizers
