```python
cols = ['primary_fuel', 'other_fuel1', 'other_fuel2', 'other_fuel3']

# Find all unique fuel names across the 4 columns (drop NaN)
all_fuels = (
    pd.unique(pd.concat([df[c] for c in cols], ignore_index=True))
)
all_fuels = [f for f in all_fuels if pd.notna(f)]

# Build one-hot columns: 1 if the fuel appears in any of the 4 columns
for fuel in all_fuels:
    df[f'fuel_{fuel}'] = (df[cols] == fuel).any(axis=1).astype(int)

df[[col for col in df.columns if col.startswith('fuel_')]]
```

# Exploratory Data Analysis

### 1. Utilize Pandas to summarize key statistics (mean, median, standard deviation) for numerical columns.

```python
# Calculate mean for each numeric column
average = df[numeric_cols].mean()
print(f"\nMean per column:\n",round(average,2))
```

```python
# Calculate median for each numeric column
median = df[numeric_cols].median()
print(f"\nMedian per column:\n",round(median,2))
```

```python
# Calculate std for each numeric column
std = df[numeric_cols].std()
print(f"\nStandard deviation per column:\n",round(std,2))
```

### 2.Explore the distribution of power plants by country and fuel type.

```python
# Count plants per country
import pandas as pd
import matplotlib.pyplot as plt

# pick the country column you have
country_col = 'country_long' if 'country_long' in df.columns else 'country'

counts = df[country_col].value_counts()           # counts per country
pct = (counts / counts.sum()).mul(100).round(1)   # optional: percentages
```

```python
# Top 20 countries by number of plants
top_n = 20
ax = counts.head(top_n).sort_values().plot(kind='barh', figsize=(7,9))
ax.set_xlabel('Number of power plants')
ax.set_ylabel('Country')
ax.set_title(f'Top {top_n} countries by number of power plants')

# add value labels
for p in ax.patches:
    ax.annotate(int(p.get_width()),
                (p.get_x() + p.get_width(), p.get_y() + p.get_height()/2),
                xytext=(5, 0), textcoords='offset points', va='center')
plt.tight_layout()
plt.show()
```

```python
# Distribution of plants based on primary fuel
primary_counts = (
    df['primary_fuel']
    .dropna()
    .value_counts()
    .sort_values(ascending=False)
)

ax = primary_counts.plot(kind='bar', figsize=(10,4),rot = 45)
ax.set_xlabel('Fuel type')
ax.set_ylabel('Number of plants')
ax.set_title('Plants by PRIMARY fuel')
plt.tight_layout()
plt.show()

# % of plants based on primary fuel 
(primary_counts / len(df) * 100).round(1).head(10)
```
