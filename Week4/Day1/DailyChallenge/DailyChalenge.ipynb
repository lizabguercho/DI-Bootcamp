```python
import numpy as np
# to generate random temparture for 12 months
temp_month = np.random.uniform(-5, 35, (10, 12)).round(0).astype(int)
```

```python
import pandas as pd
# generate names of 12 months
months = pd.date_range(start="2024-01-01", periods=12, freq="MS").strftime("%b").tolist()
# generate 10 cities
cities = pd.Series([f"City{i}" for i in range(1, 11)])
# create a data frame using months and temperatures
temp_data = pd.DataFrame(temp_month, columns=months,index = cities)
print(temp_data)
```

```python
# determine average temparature per city
annual_av_temp_per_city = temp_data.mean(axis = 1).astype(int)
annual_av_temp_per_city
```

```python
# determine the city with highest average temperature
high_temp_city = annual_av_temp_per_city.idxmax()
print(f"{high_temp_city} has the highest average temperature for the year.")
# determine the city with lowest average temperature
low_temp_city = annual_av_temp_per_city.idxmin()
print(f"{low_temp_city} has the lowest average temperature for the year.")
```

```python

import matplotlib.pyplot as plt

ax = temp_data.T.plot(figsize=(10,6))

ax.set_title("Monthly Temperatures per City")
ax.set_xlabel("Month")
ax.set_ylabel("Temperature (°C)")

ax.set_xticks(range(len(temp_data.columns)))  
ax.set_xticklabels(temp_data.columns)  

ax.legend(title="City", bbox_to_anchor=(1.05, 1), loc="upper left")
ax.grid(True)

plt.tight_layout()
plt.show()
```
