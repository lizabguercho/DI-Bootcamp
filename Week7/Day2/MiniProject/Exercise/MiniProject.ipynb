```python

def run_sql(sql,head=10):
    """Executes SQL and outputs the result as a table."""
    #Returns the entire DataFrame res so that you can continue to use the result in Python (for example, save it to CSV, filter, analyze, etc.).
    with engine.connect() as conn:
        res = pd.read_sql_query(sql, conn) #returns the result as a DataFrame (pandas table).
   
    print("\nðŸ”¸ Result:")
    #(res.head(head)) #The display() function is used in Google Colab/Jupyter to beautifully display the DataFrame as a table.
    return res.head(head) 

def exec_sql(sql):
    with engine.begin() as conn:
        conn.execute(text(sql))
    print(" SQL command executed successfully.")
```

```python
# Checkingh for NULLS
import pandas as pd

def check_nulls(table_name: str):
    print(f"\n Checking missing values in: {table_name}")

    # Get column names via LIMIT 0
    cols_df = pd.read_sql_query(f'SELECT * FROM "{table_name}" LIMIT 0;', engine)
    columns = cols_df.columns.tolist()
    if not columns:
        print("(No columns found)")
        return

    results = []
    for col in columns:
        # Quote identifiers to avoid issues with special chars / reserved words
        q = f'''
        SELECT 
            COUNT(*) AS total_rows,
            SUM(CASE WHEN "{col}" IS NULL THEN 1 ELSE 0 END) AS null_count
        FROM "{table_name}";
        '''
        res = pd.read_sql_query(q, engine)
        res["column_name"] = col
        results.append(res)

    report = pd.concat(results, ignore_index=True)
    display(
        report[["column_name", "total_rows", "null_count"]]
        .sort_values("null_count", ascending=False)
        .reset_index(drop=True)
    )
```

```python
def check_duplicates(table, key_columns):
    """Check how many duplicates exist for a table given its key column(s)."""
    key_expr = ", ".join(key_columns)
    print(f"\n Checking duplicates in {table} by {key_expr}...")
    query = f"""
    SELECT {key_expr}, COUNT(*) AS n
    FROM {table}
    GROUP BY {key_expr}
    HAVING COUNT(*) > 1
    ORDER BY n DESC
    LIMIT 10;
    """
    
    result = run_sql(query)
    if result is None or getattr(result,"empty",True):
        print("No duplicates found")
    else:
        print(result)

```

```python
run_sql("""
WITH items_per_order AS (
    SELECT 
        oi.order_id,
        SUM(oi.price + oi.freight_value) AS order_total
    FROM olist_order_items oi
    WHERE oi.price IS NOT NULL
      AND oi.freight_value IS NOT NULL
      AND oi.price > 0
      AND oi.freight_value >= 0
    GROUP BY oi.order_id
),
orders_with_value AS (
    SELECT
        o.order_id,
        o.customer_id,
        STRFTIME('%Y', o.order_purchase_timestamp) AS purchase_year,
        i.order_total
    FROM olist_orders o
    JOIN items_per_order i
      ON i.order_id = o.order_id
),
customer_year AS (
    SELECT
        purchase_year,
        customer_id,
        SUM(order_total) AS customer_year_spend,
        COUNT(*)        AS customer_year_orders
    FROM orders_with_value
    GROUP BY purchase_year, customer_id
)
SELECT
    purchase_year,
    SUM(customer_year_spend)                                   AS total_revenue,
    COUNT(DISTINCT customer_id)                                AS active_customers,
    ROUND(AVG(customer_year_spend), 2)                         AS avg_spend_per_active_customer
FROM customer_year
GROUP BY purchase_year
ORDER BY purchase_year;
""")
```

```python
run_sql("""
WITH items_per_order_seller AS (
  -- 1) Revenue contributed by each seller within each order
  SELECT
    oi.order_id,
    oi.seller_id,
    SUM(oi.price + oi.freight_value) AS seller_order_total
  FROM olist_order_items oi
  WHERE oi.price IS NOT NULL
    AND oi.freight_value IS NOT NULL
    AND oi.price > 0
    AND oi.freight_value >= 0
  GROUP BY oi.order_id, oi.seller_id
),
orders_in_window AS (
  -- 2) Only orders purchased between 2016-01-01 and 2018-12-31
  SELECT o.order_id
  FROM olist_orders o
  WHERE DATE(o.order_purchase_timestamp) BETWEEN '2016-01-01' AND '2018-12-31'
),
seller_city_revenue AS (
  -- 3) Attach seller city/state and sum revenue per seller (across their orders)
  SELECT
    s.seller_city,
    s.seller_state,
    ios.seller_id,
    SUM(ios.seller_order_total) AS seller_revenue
  FROM items_per_order_seller ios
  JOIN orders_in_window w   ON w.order_id = ios.order_id
  JOIN olist_sellers s      ON s.seller_id = ios.seller_id
  GROUP BY s.seller_city, s.seller_state, ios.seller_id
),
city_totals AS (
  -- 4) Roll up all sellers in the same city/state
  SELECT
    LOWER(TRIM(seller_city))  AS city,
    UPPER(TRIM(seller_state)) AS state,
    SUM(seller_revenue)       AS total_revenue
  FROM seller_city_revenue
  GROUP BY LOWER(TRIM(seller_city)), UPPER(TRIM(seller_state))
)
-- Top 5 cities by revenue
SELECT city, state, ROUND(total_revenue, 2) AS total_revenue
FROM city_totals
ORDER BY total_revenue DESC
LIMIT 5;
""")
```

```python
run_sql("""
        -- 1) Get one lat/lng per CUSTOMER (average per zip prefix)
WITH cust_geo AS (
  SELECT
    c.customer_id,
    c.customer_city,
    c.customer_state,
    AVG(g.geolocation_lat)  AS cust_lat,
    AVG(g.geolocation_lng)  AS cust_lng
  FROM olist_customers c
  JOIN olist_geolocation g
    ON g.geolocation_zip_code_prefix = c.customer_zip_code_prefix
  GROUP BY c.customer_id, c.customer_city, c.customer_state
),

-- 2) Get one lat/lng per SELLER (average per zip prefix)
sell_geo AS (
  SELECT
    s.seller_id,
    s.seller_city,
    s.seller_state,
    AVG(g.geolocation_lat)  AS sell_lat,
    AVG(g.geolocation_lng)  AS sell_lng
  FROM olist_sellers s
  JOIN olist_geolocation g
    ON g.geolocation_zip_code_prefix = s.seller_zip_code_prefix
  GROUP BY s.seller_id, s.seller_city, s.seller_state
),

-- 3) One row per (order, seller) to avoid double-counting items
order_seller AS (
  SELECT DISTINCT oi.order_id, oi.seller_id
  FROM olist_order_items oi
)

-- 4) Distance per order between customer city and seller city
SELECT
  cg.customer_state,
  cg.customer_city,
  sg.seller_state,
  sg.seller_city,

  -- Haversine distance (km)
  ROUND(
    6371 * 2 * ASIN(
      SQRT(
        POWER(SIN(((cg.cust_lat - sg.sell_lat) * PI() / 180) / 2), 2) +
        COS(cg.cust_lat * PI() / 180) * COS(sg.sell_lat * PI() / 180) *
        POWER(SIN(((cg.cust_lng - sg.sell_lng) * PI() / 180) / 2), 2)
      )
    )
  , 2) AS distance_km
FROM order_seller os
JOIN olist_orders o   ON o.order_id = os.order_id
JOIN cust_geo cg      ON cg.customer_id = o.customer_id
JOIN sell_geo sg      ON sg.seller_id = os.seller_id
ORDER BY distance_km DESC
LIMIT 50; """)

```
