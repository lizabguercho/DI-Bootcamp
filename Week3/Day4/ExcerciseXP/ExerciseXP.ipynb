Data visualization helps to convert raw data into visuals which are easy to interpret and draw valuable insights. We can identify trends, patterns and anomalies.

Line graphs help to visualize data changes over time. For example, temperature changes over a week or month.

Exercise 2: Creating a Line Plot for Temperature Variation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generating line plot
temp = [72, 74, 76, 80, 82, 78, 75]
days = ["Sun","Mon","Tue","Wed","Thu","Fri","Sat"]
plt.plot(days,temp,color = "purple", linewidth=2, marker ="o")
plt.xlabel("Day")
plt.ylabel("Temperature (F)")
plt.title("Temperature variations over a week")
plt.show()
```

Exercise 3: Visualizing Monthly Sales with a Bar Chart

```python
#Generating bar plot
sales = [5000, 5500, 6200, 7000, 7500, 8700,9700,10300,8500,5800,7200,7300]
month_names = list(pd.date_range(start='2025-01-01', periods=12, freq='ME').strftime('%b'))
print(month_names)
plt.bar(month_names,sales, color = "Lightblue", edgecolor="black")
plt.xlabel("Month")
plt.ylabel("Sales Amount ($)")
plt.title("Monthly sales")
```

Exercise 5: Comparing Anxiety Levels Across Different Genders

```python
mental_data.head()
```

```python
# to check if we have 2 categories in gender (F/M)
mental_data.nunique()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Filter two columns with groupby and normalize in order to get proportion of students who have anxiety by gender
proportion_data = mental_data.groupby("Choose your gender")["Do you have Anxiety?"].value_counts(normalize = True).rename("Proportion").reset_index()
```

```python

print(proportion_data)
```

```python
sns.barplot(data = proportion_data,x = "Choose your gender", y = "Proportion", hue = "Do you have Anxiety?", palette = "Purples")
plt.title("Proportion of Students Experiencing Anxiety by Gender")
plt.xlabel("Gender")
plt.ylabel("Proportion")
plt.ylim(0,1)  # because proportions go from 0 to 1
plt.show()
```

 Exercise 6: Exploring the Relationship Between Age and Panic Attacks

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

```

```python
mental_data.head(10)
```

```python
# checking unique values in "Age". Since we have nan -> we need to remove them.
print(mental_data["Age"].unique())
print(mental_data["Age"].dtype)
```

```python
# Dropping all na values
mental_data = mental_data.dropna(subset=["Age"])
```

```python
print(mental_data["Age"].isna().sum())
```

```python
print(mental_data["Age"].unique())
```

```python
# Map responses in column "Do you have panic attack" as (Yes =1, No = 0)
mental_data['panic_attack_int'] = (mental_data['Do you have Panic attack?'] == "Yes").astype(int)
display(mental_data[['Do you have Panic attack?', 'panic_attack_int']].head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(
    data=mental_data,
    x="Age",
    y="panic_attack_int",
    hue="panic_attack_int",   
    style="panic_attack_int", 
    palette="Set1",
    alpha = 0.7)

sns.stripplot(
    data=mental_data,
    x="Age",
    y="panic_attack_int",
    hue="panic_attack_int",
    palette="Set1",
    dodge=True,
    jitter=0.2,       
    alpha=0.6)

plt.yticks([0, 1], ["No", "Yes"]) 
plt.title("Relationship Between Age and Panic Attacks")
plt.xlabel("Age")
plt.ylabel("Panic Attack (0 = No, 1 = Yes)")
plt.show()
```
