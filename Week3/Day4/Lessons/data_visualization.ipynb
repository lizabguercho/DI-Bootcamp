```python
import pandas as pd
titanic_data = pd.read_csv("train.csv")
```

```python
# bar chart - each one is a distinct category
plt.bar(list(titanic_data["Survived"].unique()),titanic_data["Survived"].value_counts())
plt.show()
```

```python
plt.bar(list(titanic_data["Sex"].unique()),titanic_data["Sex"].value_counts(),color = ["red","green"])
plt.title("Count of People on the Titanic by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```
