Columns: 
'index'                     - qualitative   - It’s just an identifier
'Year'                      - quantitative  - That's a number that represents a measurable value
'Period'                    - qualitative   - Data represented as a string (Annual)
'Avg hrs per day sleeping'  - quantitative  - Represents hours of sleep,measurable value
'Standard Error'            - quantitative  - Statistical value used to measure uncertainty(numeric)
'Type of Days'              - qualitative   - Categorical data ("Weekday", "Weekend")
'Age Group'                 - qualitative   - Categories "15-24", "15 and over"
'Activity'                  - qualitative   - Data describes activities (text labels)
'Sex'                       - qualitative   - Data is represented by a category (Male,Female,Both)


Interesting columns for analysis:

1. Year -> to see if Americans are sleeping more or less over time.

2. Avg hrs per day sleeping -> The main dependent variable; useful for distribution plots, mean/median comparisons, or correlation with other factors.

3. Sex -> Allows comparison of sleep patterns between men and women.

4.Age -> Group → To analyze whether younger vs. older adults differ in sleep duration.
